## From the data we notice that:
1. Quite a lot of null values are present in almost all attributes.
2. We can deal with this null data by imputing the data with the median value for the respective attribute based on Country name
3. For attributes with few null values and weak correlation, we can drop the records as long as the number of records are very less

```python
happiness_copy2.dropna(subset=['Negative affect'],inplace=True)
```

```python
happiness_copy2.isnull().sum()
```

```python
happiness_copy2['Perceptions of corruption'] = happiness_copy2.groupby('Country name')['Perceptions of corruption'].transform(lambda x: x.fillna(x.median()))
```

```python
happiness_copy2.isnull().sum()
```

```python
happiness_copy2['Generosity'] = happiness_copy2.groupby('Country name')['Generosity'].transform(lambda x: x.fillna(x.median()))
```

```python
happiness_copy2.isnull().sum()
```

```python
happiness_copy2['Healthy life expectancy at birth'] = happiness_copy2.groupby('Country name')['Healthy life expectancy at birth'].transform(lambda x: x.fillna(x.median()))
```

```python
happiness_copy2.isnull().sum()
```

```python
happiness_copy2['Log GDP per capita'] = happiness_copy2.groupby('Country name')['Log GDP per capita'].transform(lambda x: x.fillna(x.median()))
```

```python
happiness_copy2.isnull().sum()
```

```python
happiness_copy2['Freedom to make life choices'] = happiness_copy2.groupby('Country name')['Freedom to make life choices'].transform(lambda x: x.fillna(x.median()))
```

```python
happiness_copy2.isnull().sum()
```

```python
happiness_copy2['Social support'] = happiness_copy2.groupby('Country name')['Social support'].transform(lambda x: x.fillna(x.median()))
```

```python
happiness_copy2.isnull().sum()
```

### We notice that there are some countries who have all null values for some attributes and in this case, we cannot replace value with median grouped by country

```python
happiness_copy2.corrwith(happiness["Life Ladder"]).sort_values(ascending=False)
```

### As we cannot delete these null records as doing so will delete all records of those countries, we replace these null values with median of entire attribute

```python
happiness_copy2['Healthy life expectancy at birth'].fillna(happiness_copy2['Healthy life expectancy at birth'].median(),inplace=True)
```

```python
happiness_copy2[happiness_copy2['Country name']=='Kosovo']
```

```python
happiness_copy2['Generosity'].fillna(happiness_copy2['Generosity'].median(),inplace=True)
```

```python
happiness_copy2['Log GDP per capita'].fillna(happiness_copy2['Log GDP per capita'].median(),inplace=True)
```

```python
happiness_copy2['Perceptions of corruption'].fillna(happiness_copy2['Perceptions of corruption'].median(),inplace=True)
```

```python
happiness_copy2['Social support'].fillna(happiness_copy2['Social support'].median(),inplace=True)
```

```python
happiness_copy2['Positive affect'].fillna(happiness_copy2['Positive affect'].median(),inplace=True)
```

### Here we see after all the operations, we get the dataset without any null values

```python
happiness_copy2.isnull().sum()
```

```python
happiness_copy2 = happiness_copy2[['Country name', 'year', 'Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Positive affect', 'Negative affect','Life Ladder']]
```

```python
happiness_copy2
```

### No need to scale the attributes because none of the attribute values were extremely large in absolute terms

### For predicting the Life Ladder, we won't need Year,Country name, Negative affect and Perceptions of Corruption so we are not considering those attributes. We will be needing  'Log GDP per capita','Social support', 'Healthy life expectancy at birth','Freedom to make life choices','Positive affect' as all are strong positively correlated with Life Ladder

```python
x = happiness_copy2[['Log GDP per capita','Social support', 'Healthy life expectancy at birth','Freedom to make life choices','Positive affect']]
y = happiness_copy2[['Life Ladder']]
```

### We split the data randomly into training data of 80% data and test data of 20% data

```python
from sklearn.model_selection import train_test_split
```

```python
train_set_x, test_set_x ,train_test_y, test_set_y = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
train_set_x
```

```python
test_set_x
```

```python
train_test_y
```

```python
test_set_y
```

### Creating Linear Regression model to fit our data

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(train_set_x,train_test_y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
test_set_pred = lm.predict(test_set_x)
```

### Checking for different evaluation metrics for our model

```python
from sklearn import metrics
```

```python
rmse = np.sqrt(metrics.mean_squared_error(test_set_y,test_set_pred))
```

```python
rmse
```

```python
metrics.mean_absolute_error(test_set_y,test_set_pred)
```

```python
metrics.mean_squared_error(test_set_y,test_set_pred)
```

```python
test_set_pred
```

```python
plt.scatter(test_set_y,test_set_pred)
```
